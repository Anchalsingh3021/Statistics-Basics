1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

  -  Descriptive statistics are used to describe and summarize the information we already have. For example, calculating the average marks of a class or making a bar graph of students’ test scores.
Inferential statistics go a step further and help us make predictions or conclusions about a larger group based on a smaller sample. For example, studying the marks of 30 students and predicting how well the whole school might perform.
So, descriptive statistics tell us “what is happening,” while inferential statistics help us guess or conclude “what might be true beyond the data we have.”


2. What is sampling in statistics? Explain the differences between random
and stratified sampling.
  
  - Sampling in statistics means selecting a small group from a large population to study, instead of collecting data from everyone. Random sampling is when every person has an equal chance of being chosen, like picking names from a hat. Stratified sampling is when the population is divided into groups (like boys and girls or age groups), and samples are taken from each group. Random sampling is simpler but may not represent all groups well. Stratified sampling gives a better balance because each group is included.

3. Define mean, median, and mode. Explain why these measures of central
tendency are important.

  - Mean, median, and mode are three ways to find the center or average value of data. The **mean** is the average, found by adding all values and dividing by the number of values. The **median** is the middle value when the data is arranged in order. The **mode** is the value that appears the most in the data. These measures are important because they help us understand what a “typical” value looks like in a group of numbers and make data easier to compare and explain.


4. Explain skewness and kurtosis. What does a positive skew imply about the
data?

  - Skewness tells us whether the data is spread more to one side of the average, while kurtosis explains how sharp or flat the peak of the data distribution is. If data has **positive skew**, it means the long tail is on the right side. This shows that most values are on the lower side, and a few large values pull the average to the right. In simple words, positive skew means more small values and fewer very big values. These ideas help us understand the shape and behavior of data.




```python
"""5. Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]"""



numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculating mean
mean = sum(numbers) / len(numbers)

# Calculating median
numbers_sorted = sorted(numbers)
mid = len(numbers_sorted) // 2
if len(numbers_sorted) % 2 == 0:
    median = (numbers_sorted[mid - 1] + numbers_sorted[mid]) / 2
else:
    median = numbers_sorted[mid]

# Calculating mode
from statistics import mode
mode_value = mode(numbers)

print("Mean:", mean)
print("Median:", median)
print("Mode:", mode_value)
```

```python
"""6. Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]"""



list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculating covariance
mean_x = sum(list_x) / len(list_x)
mean_y = sum(list_y) / len(list_y)

covariance = sum((x - mean_x) * (y - mean_y) for x, y in zip(list_x, list_y)) / len(list_x)

# Calculating correlation coefficient
import math
std_x = math.sqrt(sum((x - mean_x) ** 2 for x in list_x) / len(list_x))
std_y = math.sqrt(sum((y - mean_y) ** 2 for y in list_y) / len(list_y))

correlation = covariance / (std_x * std_y)

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
"""7. Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
"""

import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data)
plt.title("Boxplot of Data")
plt.show()

# Finding outliers using IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_limit or x > upper_limit]

print("Outliers:", outliers)
```

```python
"""8. You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]"""

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Calculate means
mean_ad = sum(advertising_spend) / len(advertising_spend)
mean_sales = sum(daily_sales) / len(daily_sales)

# Calculate covariance
covariance = sum((a - mean_ad) * (s - mean_sales)
                 for a, s in zip(advertising_spend, daily_sales)) / len(advertising_spend)

# Calculate correlation
import math
std_ad = math.sqrt(sum((a - mean_ad)**2 for a in advertising_spend) / len(advertising_spend))
std_sales = math.sqrt(sum((s - mean_sales)**2 for s in daily_sales) / len(daily_sales))

correlation = covariance / (std_ad * std_sales)

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)


```
